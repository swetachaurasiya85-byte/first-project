```python
def add(x,y):
    return x+y
    
def sub(x,y):
    return x-y

def multiply(x,y):
    return x*y

def divide(x,y):
    if y==0:
       return "error it is not divisible"
    else:
        return x/y 

print("select the poeration")
print("1. addition")
print("2. subtraction ")
print("3. multiply")
print("4. division")

choice=input("enter the choice 1/2/3/4:")

num1 =float(input("enter the num1:"))
num2= float(input("enter the num2:"))

if choice=='1':
    print(f"the addition of number is {add(num1,num2)}")
elif choice=='2':
    print(f"the subtaction of number is {sub(num1,num2)}")
elif choice=='3':
    print(f"the multiplication of number is {multiply(num1,num2)}")
elif choice=='4':
    print(f"the division of number is {divide(num1,num2)}")
else:
    print("invalid input please enter valid option")




```
